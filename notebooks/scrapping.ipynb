```python
years=["2016","2013","2007","2002","1998","1995"]
```

```python
url_list=[]

for year in years:
    url_list.append(f"https://africanelephantdatabase.org/report/{year}/Africa")

url_list
```

```python
# sacamos la tabla para cada una de las url de arriba

driver = webdriver.Chrome(ChromeDriverManager().install())

result_list=[]

for i in url_list:
    try:
        driver.get(i)
        result=driver.find_element("xpath", '//*[@id="tab-section-data-quality-page-0"]').text
        result_list.append(result)
    except:
        try:
            driver.get(i)
            result=driver.find_element("xpath", '/html/body/div[3]/div/div/table[4]').text
            result_list.append(result)
        except:
            driver.get(i)
            result=driver.find_element("xpath", '/html/body/div[3]/div/div/table[2]').text
            result_list.append(result)

driver.quit()
```
